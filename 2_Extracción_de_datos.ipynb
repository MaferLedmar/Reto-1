```python
	
#Importamos la librería: pandas
import pandas as pd 
import io 

#Leemos el contenido del archivo
#df = pd.read_csv(io.BytesIO(uploaded['file.csv']), index_col=2) 
df = pd.read_csv('Ventas_totales_supermercados.csv') 
df.info()
df

#Importar archivo desde Github
#url = 'copied_raw_github_link'
#df = pd.read_csv(url)
```

```python
#Filtro 1
#No es posible aplicar el filtro por variables cualitativas
```

```python
#Filtro por filas
filtro2= df.iloc[30:32,: ] #Para filas consecutivas

#filtro2=df.iloc[ [10, 20, 30] , : ]  #filas no consecutivas 
filtro2

#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2.csv")
```

```python
#Filtro por columnas
#filtro3=df.iloc[ : , 4:7] #Para columnas consecutivas
filtro3=df.iloc[ : , [1,5,11,14]]  #columnas no consecutivas 
filtro3

#Convertir archivo filtrado a CSV
filtro3.to_csv("Filtro3.csv")
```

```python
#Filtro por cabecera
filtro5=df.head()
filtro5

#Convertir archivo filtrado a CSV
filtro5.to_csv("Filtro5.csv")
```

```python
#FILTROS BASICOS PARA VARIABLES CUANTITATIVAS
#Filtro por comparacion
filtro6=df[df["ventas_precios_corrientes"] > 70000]
filtro6
```

```python
#Convertir archivo filtrado a CSV
filtro6.to_csv("Filtro6.csv")
```

```python
#FILTROS BASICOS CON CONETORES LOGICOS
#Filtro Y
filtro7=df[(df["ventas_precios_corrientes"] > 100000) & (df["indice_tiempo"] == "2022-01-01")]
filtro7

#Convertir archivo filtrado a CSV
filtro7.to_csv("Filtro7.csv")
```

```python
#Filtro o
filtro8=df[(df["indice_tiempo"] == "2018-05-01") | (df["ventas_precios_corrientes"] > 100000)]
filtro8

#Convertir archivo filtrado a CSV
filtro8.to_csv("Filtro8.csv")
```

```python
#Filtro not
filtro9=df[~(df["ventas_precios_corrientes"] >= 100000) & (df["indice_tiempo"] == "2020-06-01")]
filtro9

#Convertir archivo filtrado a CSV
filtro9.to_csv("Filtro9.csv")
```

```python
#Filtro de filas vs columnas 

# Para aplicar este filtro SI se requiere tener una columna como índice
### df1 = pd.read_csv(io.BytesIO(uploaded['Ventas_totales_supermercados.csv']), index_col=0)

df1 = pd.read_csv('Ventas_totales_supermercados.csv', index_col=0)

filtro4=df1.loc[["2017-01-01","2022-01-01"], ["ventas_precios_corrientes", "ventas_precios_constantes"]]
filtro4
```

```python
#Convertir archivo filtrado a CSV
filtro4.to_csv("Filtro4.csv")
```
